```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('crayons2.jpg', cv2.IMREAD_COLOR)



rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imshow("im",img)
#cv2.imshow("rgb_image",rgb_image)


red_channel = rgb_image[:, :, 0]
green_channel = rgb_image[:, :, 1]
blue_channel = rgb_image[:, : , 2]



# Transformando a HSV

hsv_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)

cv2.imshow("hsv_image",hsv_image)


hue_channel = hsv_image[:, :, 0]
saturation_channel = hsv_image[:, :, 1]
value_channel = hsv_image[:, :, 2]


cv2.imshow("hue_channel",hue_channel)
cv2.imshow("saturation_channel",saturation_channel)
cv2.imshow("value_channel",value_channel)


ama_bajo = np.array([15,0,0])
ama_alto = np.array([36,255,255])
#rojo_bajo = np.array([155,155,84])
#rojo_alto = np.array([180,255,255])


mascara_ama = cv2.inRange(hsv_image , ama_bajo, ama_alto )

imFiltrada = cv2.bitwise_and(img, img, mask=mascara_ama );
cv2.imshow( "filtrado Amarillos", mascara_ama)
cv2.imshow( "filtrado imFiltrada", imFiltrada)




cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)

```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('flor.jpg', cv2.IMREAD_COLOR)

rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)

cv2.imshow("im",img)

ama_bajo = np.array([15,0,0])
ama_alto = np.array([36,255,255])

verde_bajo = np.array([36,0,0])
verde_alto = np.array([70,255,255])



mascara_ama = cv2.inRange(hsv_image , ama_bajo, ama_alto )
mascara_verde = cv2.inRange(hsv_image , verde_bajo, verde_alto )


mascara = cv2.bitwise_or(mascara_ama, mascara_verde)

imFiltrada = cv2.bitwise_and(img, img, mask=mascara );

cv2.imshow( "filtrado imFiltrada", imFiltrada)




cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
